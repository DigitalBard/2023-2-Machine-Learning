```python
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
fmnist = keras.datasets.fashion_mnist

(x_train, y_train), (x_test, y_test) = fmnist.load_data()
```

## 문제 1

```python
model = keras.models.Sequential([
    keras.layers.Conv2D(input_shape=(28, 28, 1), activation='relu',
                        kernel_size=(3, 3), filters=128),
    keras.layers.MaxPooling2D((2, 2), strides=2),
    keras.layers.Flatten(),
    keras.layers.Dense((14 * 14), activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
_hist = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

df = _hist.history
```

## 문제 2

```python
norm_x_train = x_train / 255
norm_x_test = x_test / 255
```

```python
norm_model = keras.models.Sequential([
    keras.layers.Conv2D(input_shape=(28, 28, 1), activation='relu',
                        kernel_size=(3, 3), filters=128),
    keras.layers.MaxPooling2D((2, 2), strides=2),
    keras.layers.Flatten(),
    keras.layers.Dense((14 * 14), activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

norm_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
norm_hist = norm_model.fit(norm_x_train, y_train, epochs=10, validation_data=(norm_x_test, y_test))

norm_df = norm_hist.history
```

## 문제 3

```python
my_model = keras.models.Sequential([
    keras.layers.Conv2D(input_shape=(28, 28, 1), activation='relu',
                        kernel_size=(3, 3), padding='same', filters=64),
    keras.layers.MaxPooling2D((2, 2), strides=2),
    keras.layers.Dropout(rate=0.2),
    keras.layers.Conv2D(activation='relu', kernel_size=(3, 3), filters=64),
    keras.layers.MaxPooling2D((2, 2), strides=2),
    keras.layers.Dropout(rate=0.2),
    keras.layers.Flatten(),
    keras.layers.Dense((14 * 14), activation='relu'),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

my_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
my_hist = my_model.fit(norm_x_train, y_train, epochs=10, validation_data=(norm_x_test, y_test))

my_df = my_hist.history
```

## 문제 4

```python
# accuracy 비교
plt.plot([i for i in range(10)], df['accuracy'], color='red', linestyle='-', label='base')
plt.plot([i for i in range(10)], norm_df['accuracy'], color='blue', linestyle='-', label='base + norm')
plt.plot([i for i in range(10)], my_df['accuracy'], color='black', linestyle='-', label='layer + norm')
plt.title('accuracy')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('accuracy')
print()
```

```python
# validation accuracy 비교
plt.plot([i for i in range(10)], df['val_accuracy'], color='red', linestyle='-', label='base')
plt.plot([i for i in range(10)], norm_df['val_accuracy'], color='blue', linestyle='-', label='base + norm')
plt.plot([i for i in range(10)], my_df['val_accuracy'], color='black', linestyle='-', label='layer + norm')
plt.title('val_accuracy')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('val_accuracy')
print()
```
