## 문제 2

```python
def spam_classifier():
  train_file = path+"lab5_train.csv"
  train_data = pd.read_csv(train_file).values

  data_x = train_data[:, :-1]
  data_y = train_data[:, -1]

  svc = SVC(C=1, kernel='linear')
  svc.fit(data_x, data_y)

  return svc
```

## 문제 3

```python
def test_spam_classifier(test_data):
  model = spam_classifier()

  x_test = test_data[:,:-1]
  y_test = test_data[:,-1]

  print("테스트 셀의 정확도: {:.3f}".format(model.score(x_test, y_test)))
```

```python
test_file = path+"lab5_test.csv"
test_data = pd.read_csv(test_file).values

test_spam_classifier(test_data)
```
