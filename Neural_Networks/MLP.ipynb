```python
from sklearn.neural_network import MLPClassifier
import numpy as np
```

```python
# XOR 학습 데이터 구성

x = np.array([[0, 0], [1, 1], [0, 1], [1, 0]])
y = np.array([0, 0, 1, 1])
```

```python
model = MLPClassifier( max_iter=100,
                       solver='lbfgs',
                       hidden_layer_sizes=(),
                       random_state=0)
model.fit(x, y)
y_hat = model.predict(x)

print(f'Weights : {model.coefs_}')
print(f'Bias : {model.intercepts_}')

print()
print('{:^40}'.format(' Predicted Result '))
print("-----------------------------------------")
print("|     x1    |     x2    |  class label  |")
print("-----------------------------------------")
for i in range(len(x)):
  print(f'|{x[i][0]:^11}|{x[i][1]:^11}|{y_hat[i]:^15}|')
```

```python
def sigmoid(x):
  result = 1 / (1+np.exp(-x))
  return result
```

```python
def print_output_of_hidden_layer(x, weights, bias):
  activation_input = np.dot(x, weights) + np.array([bias.tolist()] * len(x))
  return sigmoid(activation_input)
```

```python
model = MLPClassifier( max_iter=100,
                       activation='logistic',
                       solver='lbfgs',
                       hidden_layer_sizes=(2, ),
                       random_state=0)
model.fit(x, y)
y_hat = model.predict(x)

weights = model.coefs_
bias = model.intercepts_
print(f'Weights : {weights}')
print(f'Bias : {bias}')
print()

activation_output = print_output_of_hidden_layer(x, weights[0], bias[0])
print('output from hidden layer : ')
print(f'{activation_output}')
print()

print()
print('{:^40}'.format(' Predicted Result '))
print("-----------------------------------------")
print("|     x1    |     x2    |  class label  |")
print("-----------------------------------------")
for i in range(len(x)):
  print(f'|{x[i][0]:^11}|{x[i][1]:^11}|{y_hat[i]:^15}|')
```
