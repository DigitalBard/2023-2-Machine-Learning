```python
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
fmnist = keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = fmnist.load_data()
```

## 문제 1

```python
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape = (28, 28)),
    keras.layers.Dense(1024, activation='relu'),
    keras.layers.Dense(512, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
hist = model.fit(x_train, y_train, epochs=10, validation_data = (x_test, y_test))
```

```python
df = hist.history
```

## 문제 2

```python
norm_x_train = x_train / 255
norm_x_test = x_test / 255
```

```python
norm_model = keras.models.Sequential([
    keras.layers.Flatten(input_shape = (28, 28)),
    keras.layers.Dense(1024, activation='relu'),
    keras.layers.Dense(512, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

norm_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
norm_hist = norm_model.fit(norm_x_train, y_train, epochs=10, validation_data = (norm_x_test, y_test))
```

```python
norm_df = norm_hist.history
```

## 문제 3

```python
expanded_model = keras.models.Sequential([
    keras.layers.Flatten(input_shape = (28, 28)),
    keras.layers.Dense(1024, activation='relu'),
    keras.layers.Dense(512, activation='relu'),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

expanded_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
expanded_hist = expanded_model.fit(norm_x_train, y_train, epochs=10, validation_data = (norm_x_test, y_test))
```

```python
expanded_df = expanded_hist.history
```

## 문제 4

```python
# accuracy 비교
plt.plot([i for i in range(10)], df['accuracy'], color='red', linestyle='-', label='base')
plt.plot([i for i in range(10)], norm_df['accuracy'], color='blue', linestyle='-', label='base + norm')
plt.plot([i for i in range(10)], expanded_df['accuracy'], color='black', linestyle='-', label='node + norm')
plt.title('accuracy')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('accuracy')
print()
```

```python
# validation accuracy 비교
plt.plot([i for i in range(10)], df['val_accuracy'], color='red', linestyle='-', label='base')
plt.plot([i for i in range(10)], norm_df['val_accuracy'], color='blue', linestyle='-', label='base + norm')
plt.plot([i for i in range(10)], expanded_df['val_accuracy'], color='black', linestyle='-', label='node + norm')
plt.title('val_accuracy')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('val_accuracy')
print()
```
