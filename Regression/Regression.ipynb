```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd
```

## 문제 1

```python
path = "./drive/MyDrive/colab/ML_lab/"
train_file = path+"lab7_train.csv"

train_data = pd.read_csv(train_file).values
xt = train_data[:, :-1]
yt = train_data[:, -1]
```

```python
lin_model = LinearRegression()
lin_model.fit(xt, yt)
y_hat = lin_model.predict(xt)

print(f"MSE LOSS: {mean_squared_error(yt, y_hat)}")
```

```python
test_file = path+"lab7_test.csv"

test_data = pd.read_csv(test_file).values
x_test = test_data[:, :-1]
y_test = test_data[:, -1]
```

```python
y_hat = lin_model.predict(x_test)

print(f"MSE LOSS: {mean_squared_error(y_test, y_hat)}")
```

## 문제 2

```python
path = "./drive/MyDrive/colab/ML_lab/"
train_file = path+"lab7_train.csv"

train_data = pd.read_csv(train_file).values
xt = train_data[:, :-1]
yt = train_data[:, -1]

number_of_features = len(xt[0])
number_of_samples = len(xt)
```

```python
def h(x, param):
  return np.dot(x, param[: -1]) + param[-1]
```

```python
learning_iteration = 3000
learning_rate = 0.1

param = np.ones((number_of_features + 1), )

for i in range(learning_iteration):
  error = h(xt, param) - yt
  param[:-1] -= learning_rate * (2 / number_of_samples) * np.dot(error, xt)
  param[-1] -= learning_rate * (2 / number_of_samples) * error.sum()
```

```python
y_hat = h(xt, param)

print(f"MSE LOSS: {mean_squared_error(yt, y_hat)}")
```

```python
test_file = path+"lab7_test.csv"

test_data = pd.read_csv(test_file).values
x_test = test_data[:, :-1]
y_test = test_data[:, -1]
```

```python
y_hat = h(x_test, param)

print(f"MSE LOSS: {mean_squared_error(y_test, y_hat)}")
```
